```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../SVM/Social_Network_Ads.csv')

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.25, random_state = 0
)
```

```python
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', random_state=0)
svm.fit(X_train, y_train)
```

```python
svm.predict(sc.transform([[30, 8700]]))
```

```python
y_pred = svm.predict(X_test)


print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt


# sc.inverse_transform digunakan untuk mengembalikan nilai awal semula
X_set, y_set = sc.inverse_transform(X_train), y_train

# np.meshgrid digunakan untuk membuat grid dari dua atau lebih array 1D.

# contoh nya adalah 
#  X = np.array([1, 2, 3])
#  Y = np.array([4, 5, 6])
 
#  X, Y = np.meshgrid(X, Y)
 
#  print(X)
#  print(Y)
 
#  output : 
#  X:
# [[1 2 3]
#  [1 2 3]
#  [1 2 3]]

# Y:
# [[4 4 4]
#  [5 5 5]
#  [6 6 6]]    

# np.arange adalah fungsi untuk menghasilkan array 
# dengan nilai yang berada dalam rentang tertentu dengan langkah tertentu

# contoh : 
# np.arange(start, stop, step)
# start: Nilai awal (default adalah 0).
# stop: Nilai akhir (nilai ini tidak termasuk dalam array).
# step: Jarak antar elemen (default adalah 1).

# arr = np.arange(0, 10, 2)
# print(arr)

# output:
#     [0 2 4 6 8]


X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))


# plt.contourf adalah fungsi dari Matplotlib yang digunakan untuk menggambar kontur terisi pada plot. 
# Kontur terisi menggambarkan daerah-daerah yang memiliki nilai yang sama atau serupa dalam data 2D.

# Penggunaan : 
# plt.contourf(X, Y, Z, **kwargs), di mana:
# X: Koordinat sumbu X.
# Y: Koordinat sumbu Y.
# Z: Nilai yang digunakan untuk menentukan warna kontur.
# alpha: Transparansi area terisi (nilai antara 0 dan 1).
# cmap: Skema warna untuk kontur.


# ravel() adalah metode dari NumPy yang digunakan untuk mengubah array multi-dimensi menjadi satu dimensi.
# contoh :
# import numpy as np

# arr = np.array([[1, 2, 3], [4, 5, 6]])
# flat_arr = arr.ravel()
# print(flat_arr)
# Output: [1 2 3 4 5 6]


# reshape() adalah metode dari NumPy yang digunakan untuk mengubah bentuk (dimensi) 
# dari array tanpa mengubah datanya.

# import numpy as np
# arr = np.array([1, 2, 3, 4, 5, 6])
# reshaped_arr = arr.reshape((2, 3)) -> (dua baris, tiga kolom)
# print(reshaped_arr)

# Output:
# [[1 2 3]
#  [4 5 6]]


plt.contourf(X1, X2, svm.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'blue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())



# enumerate() adalah fungsi built-in Python yang digunakan untuk mengiterasi elemen dalam iterable 
# (seperti list atau tuple) dengan mendapatkan indeks dari elemen tersebut.

# fruits = ['apple', 'banana', 'cherry']
# for index, fruit in enumerate(fruits):
#     print(index, fruit)

# Output:
# 0 apple
# 1 banana
# 2 cherry


# np.unique() adalah fungsi dari NumPy yang digunakan untuk mengembalikan elemen unik dari array,
# serta jumlah kemunculan setiap elemen jika diinginkan.

# import numpy as np
# arr = np.array([1, 2, 2, 3, 4, 4, 4, 5])
# unique_elements = np.unique(arr)

# print(unique_elements)  
# Output: [1 2 3 4 5]
for i, j in enumerate(np.unique(y_set)):
    plt.scatter():
    # Ini adalah fungsi dari Matplotlib yang digunakan untuk membuat plot pencar (scatter plot).
    # Fungsi ini memplot titik-titik pada grafik, di mana setiap titik bisa mewakili satu data.
    # X_set[y_set == j, 0]: Ini memilih semua elemen dari X_set (data input) di mana label y_set sama dengan j, 
    # dan memilih kolom pertama (sumbu X).
    # X_set[y_set == j, 1]: Ini memilih semua elemen dari X_set di mana label y_set sama dengan j, 
    # dan memilih kolom kedua (sumbu Y).
    # c: Ini mengatur warna titik-titik berdasarkan label yang diberikan. 
    # Di sini, kita menggunakan ListedColormap untuk mendefinisikan warna titik.
    # label = j: Ini memberikan label untuk setiap kelompok data yang dipetakan, 
    # yang akan ditampilkan dalam legenda plot.
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('maroon', 'cyan'))(i), label = j)
    
plt.title('SVM (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
